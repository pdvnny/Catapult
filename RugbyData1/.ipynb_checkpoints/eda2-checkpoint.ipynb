```python
# IMPORTS
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
import seaborn as sns

# OTHER SETUP

dir_list = os.listdir("./cleaned_data")
```

### Loading All Player Data - combines all data into one Pandas DataFrame

Name of DataFrame: `all_players_df`

Documentation References:
* https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html

```python
seed = dir_list[-1]
all_players_df = pd.read_csv("cleaned_data/"+seed, header=0, index_col=0)

dir_list = dir_list[:-2]

for file in dir_list:
    new_csv = pd.read_csv("cleaned_data/"+file, header=0, index_col=0)
    all_players_df = pd.concat([all_players_df, new_csv], axis=0)
    
    
## REALISTICALLY I DON'T NEED OR WANT "first_name" AND "last_name" IN MY DATAFRAMES
all_players_df = all_players_df.drop(columns=['first_name','last_name'])
    

print(all_players_df.head(2))
#print(all_players_df.jersey.unique())

jersey_ids = all_players_df.jersey.unique()
print("\nHere are all the unique jersey numbers for the match:")
print(jersey_ids)
```

## Second Step: Create a looping process (NO! -> a function!!!) for making the plots above

INPUTS:
1. THE DATA
2. Player jersey! --- could also be any other identifier, but lets go with jersey for now

```python
def plot_player_position_data(all_players_data, jersey_id):
    # Data extraction
    player = all_players_data.loc[(all_players_data["jersey"] == jersey_id), :]
    player_h1 = player.loc[(player.period=="1st Half"),["X","Y","game_time"]]
    player_h2 = player.loc[(player.period=="2nd Half"),["X","Y","game_time"]]
    
    # Prepare plot
    sns.set_theme(style="darkgrid")
    fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)
    fig.suptitle("Player {0} - Device #: {1} - Position: {2}".format(player.jersey.iloc[0],
                                                                    player.device_id.iloc[0],
                                                                    player.position_name.iloc[0]))
    # Plotting the data
    sns.scatterplot(ax=axes[0], data=player_h1, x="X", y="Y")
    axes[0].set_title(r"$1^{st}$ Half")

    sns.scatterplot(ax=axes[1], data=player_h2, x="X", y="Y")
    axes[1].set_title(r"$2^{nd}$ Half")
    
    # Adding text to the plot
    # Grabbing the position information
    valid_position_indexes_h1 = np.isfinite(player_h1.X)
    valid_positions_h1 = player_h1.loc[valid_position_indexes_h1,["X","Y"]].reset_index(drop=True)
    h1_start_X, h1_start_Y = valid_positions_h1.iloc[0]+2
    h1_end_X, h1_end_Y = valid_positions_h1.iloc[-1]+2

    valid_position_indexes_h2 = np.isfinite(player_h2.X)
    valid_positions_h2 = player_h2.loc[valid_position_indexes_h2,["X","Y"]].reset_index(drop=True)
    h2_start_X, h2_start_Y = valid_positions_h2.iloc[0]+2
    h2_end_X, h2_end_Y = valid_positions_h2.iloc[-1]+2
    
    # Adding the labels to the plot
    font_style = {'color':'black', 'size':10}
    box_style = {'facecolor':'red', 'alpha':0.5}
    
    axes[0].text(x=h1_start_X, y=h1_start_Y, s="START", fontdict=font_style, bbox=box_style)
    axes[0].text(x=h1_end_X, y=h1_end_Y, s="END", fontdict=font_style, bbox=box_style)
    axes[1].text(x=h2_start_X, y=h2_start_Y, s="START", fontdict=font_style, bbox=box_style)
    axes[1].text(x=h2_end_X, y=h2_end_Y, s="END", fontdict=font_style, bbox=box_style)
    
# Testing the function
plot_player_position_data(all_players_df, "G23")
```
